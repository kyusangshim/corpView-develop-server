```python
import pandas as pd
```

```python
# '11차개정한국표준산업분류' 시트 불러오기
df = pd.read_excel('./한국표준산업분류(11차)표.xlsx', sheet_name='11차개정한국표준산업분류', skiprows=2)
```

```python
# 컬럼 이름 정리
df.columns = ["code_1", "name_1", "code_2", "name_2", "code_3", "name_3", "code_4", "name_4", "code_5", "name_5"]

# 상위 분류 누적 (NaN 채우기)
df.fillna(method='ffill', inplace=True)

# 코드 열들을 string으로 처리 (소수점 제거)
for col in ["code_2", "code_3", "code_4", "code_5"]:
    df[col] = df[col].apply(lambda x: str(int(x)) if pd.notnull(x) else None)
```

```python
code_columns = {
    "code_2": 2,
    "code_3": 3,
    "code_4": 4,
    "code_5": 5,
}

for col, width in code_columns.items():
    df[col] = df[col].apply(lambda x: str(x).zfill(width) if pd.notnull(x) else None)
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine("mysql+pymysql://root:1234@localhost:3310/stock_db")  # <- 환경에 맞게 수정

# 테이블에 INSERT
df.to_sql("industry_classification", con=engine, if_exists="append", index=False)
```
