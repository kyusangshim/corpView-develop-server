```python
import asyncio
import nest_asyncio
from sqlalchemy.orm import sessionmaker, Session
from sqlalchemy import create_engine
from sqlalchemy.future import select
import sys
import os
```

```python
project_root = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
if project_root not in sys.path:
    sys.path.append(project_root)

from domains.companies.model import CompanyOverviews
from services.logo_api import update_company_logo
from core.config import DATABASE_URL
```

```python
# 데이터베이스 연결 설정
engine = create_engine(DATABASE_URL)
SessionLocal = sessionmaker(autocommit=False, autoflush=False, bind=engine)

def update_all_company_logos():
    db: Session = SessionLocal()

    try:
        companies = db.query(CompanyOverviews).all()

        if not companies:
            print("회사를 찾을 수 없습니다")
            return
        
        print(f"총 {len(companies)}개의 회사 로고 업데이트를 시작합니다.")

        # 각 회사에 대해 로고 업데이트 함수를 동기적으로 실행합니다.
        for i, company in enumerate(companies):
            # update_company_logo 함수는 내부적으로 commit을 수행합니다.
            update_company_logo(company, db)
            print(f"({i + 1}/{len(companies)}) {company.corp_name} 로고 업데이트 완료.")

        print("모든 회사 로고 업데이트가 완료되었습니다.")

    finally:
        db.close()

# 스크립트 실행
if __name__ == "__main__":
    update_all_company_logos()


```
